## Plot multiple data sets on the same graph - 2

```python
# style
# line chart
# import the necessary packages and modules
import matplotlib.style  as style
import matplotlib.pyplot as plt
# use style
style.use('ggplot')
```

Prepare the Data

```python
x1 = [5,8,10,12]
y1 = [12,16,6,9]

x2 = [6,9,11,15]
y2 = [6,15,7,5]

x3 = [6,9,13,18]
y3 = [6,15,4,2]
```

---
Prepare the plot

```python
plt.figure()
plt.plot(x1,y1,'g', label='line one', marker='o', linestyle='-', linewidth=1)
plt.plot(x2,y2,'c', label='line two', marker='.', linestyle='-', linewidth=10)
plt.plot(x3,y3,'m', label='line three', marker='X', linestyle='--', linewidth=2)
```

```python
#data
x1 = [5,8,10,12]
y1 = [12,16,6,9]

x2 = [6,9,11,15]
y2 = [6,15,7,5]

x3 = [6,9,13,18]
y3 = [6,15,4,2]

#plot preparation
plt.figure()
plt.plot(x1,y1,'g', label='line one', marker='o', linestyle='-', linewidth=1)
plt.plot(x2,y2,'c', label='line two', marker='.', linestyle='-', linewidth=10)
plt.plot(x3,y3,'m', label='line three', marker='X', linestyle='--', linewidth=2)

# annotations
plt.title('Graph Title')
plt.ylabel('Y axis')
plt.xlabel('X axis')

# show plot
plt.legend()
plt.show()
```
