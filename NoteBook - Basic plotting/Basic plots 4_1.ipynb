```python
import matplotlib.pyplot as plt
#%matplotlib inline  #(alternatively)
import numpy as np
```

```python
import matplotlib.style as style
style.use('ggplot')
x1 = [5,8,10]
y1 = [12,16,6]
x2 = [6,9,11]
y2 = [6,15,7]

# prepare plot
plt.figure()
plt.bar(x1,y1, label='line one')
plt.title('Graph Title')
plt.ylabel('Y axis')
plt.xlabel('X axis')
# show plot
plt.legend()
plt.show()
```
