### Draw the cosine and sine functions on the sameplot

- First step is to get the data for the sine and cosine functions:
- import the necessary packages and modules

```python
import matplotlib.pyplot as plt
#%matplotlib inline  #(alternatively)
import numpy as np
```

```python
X = np.linspace(-np.pi, np.pi, 256) #256 no. of points, nothing special. 
C = np.cos(X)
S = np.sin(X)
```

```python
# prepare plot
plt.figure()
plt.plot(X,C)
plt.plot(X,S)
# show plot
plt.show()
```

```python
plt.figure()
plt.plot(X,C)
plt.plot(X,S,label='data')
# annotations
plt.title('Graph Title') 
plt.xlabel('X axis')
plt.ylabel('Y axis')
# add a legend
plt.legend()
```
