## Importance of Label & Legend

 `both are required if legend is to be shown`

```python
# style
# line chart
# import the necessary packages and modules
import matplotlib.style  as style
import matplotlib.pyplot as plt
# use style
style.use('ggplot')
```

## Plot multiple data sets on the same graph

Prepare the Data

```python
x1 = [5,8,10,12]
y1 = [12,16,6,9]

x2 = [6,9,11,15]
y2 = [6,15,7,5]

x3 = [6,9,13,18]
y3 = [6,15,4,2]
```

---

Prepare the plot

```python
plt.figure()
plt.plot(x1,y1,label='dataset 1')
plt.plot(x2,y2,label='dataset 2')
plt.plot(x3,y3,label='dataset 3')
```

---

Annotations

```python
plt.title('Graph Title')
plt.ylabel('Y axis')
plt.xlabel('X axis')
```

---
show plot with grids

```python
plt.grid(True,color='k')
```

### Now Plot all elemets together 

```python
#Data preparation

x1 = [5,8,10,12]
y1 = [12,16,6,9]

x2 = [6,9,11,15]
y2 = [6,15,7,5]

x3 = [6,9,13,18]
y3 = [6,15,4,2]

#Prepare the plot for above data

plt.figure()
plt.plot(x1,y1,label='dataset 1')
plt.plot(x2,y2,label='dataset 2')
plt.plot(x3,y3,label='dataset 3')

#Title and axis

plt.title('Graph Title')
plt.ylabel('Y axis')
plt.xlabel('X axis')

#add grids to the plot

plt.grid(True,color='k')

#add legends to the plot

plt.legend()

#show the plot

plt.show()
```
